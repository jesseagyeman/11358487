```python
wine_data.plot(x = 'pH' , y = 'density' , ylabel = 'density' , kind = 'scatter' , title = 'scatter plot of density against pH')
plt.savefig('plot2.png')
```

```python
quality_group = wine_data.groupby(wine_data['quality'].head(50))
total_quality = quality_group['quality'].agg(['sum'])
total_quality.plot(kind = 'bar' , title = 'Bar graph of quality')
plt.savefig('plot3.png')
```

```python
wine_data.corr()
```

```python
mean_chl = wine_data['chlorides'].mean()
mean_fa  = wine_data['fixed acidity'].mean()
mean_fsd = wine_data['free sulphur dioxide'].mean()
mean_dty = wine_data['density'].mean()
mean_ql  = wine_data['quality'].mean()
mean_va  = wine_data['volatile acidity'].mean()
```

```python
print(f' chlorides: {mean_chl}')
print(f' fixed acidity: {mean_fa}')
print(f' free sulphur dioxide: {mean_fsd}')
print(f' density: {mean_dty}')
print(f' quality: {mean_ql}')
print(f' volatile acidity: {mean_va}')
```

```python
chlorides: 0.08747586206896552
fixed acidity: 8.318335419274094
 free sulphur dioxide: 15.867334167709638
 density: 0.996745519399249
 quality: 5.635794743429287
 volatile acidity: 0.527747183979975
```

```python
wine_data['chlorides'] = wine_data['chlorides'].fillna(mean_chl)
wine_data['fixed acidity'] = wine_data['fixed acidity'].fillna(mean_fa)
wine_data['free sulphur dioxide'] = wine_data['free sulphur dioxide'].fillna(mean_fsd)
wine_data['density'] = wine_data['density'].fillna(mean_dty)
wine_data['quality'] = wine_data['quality'].fillna(mean_ql)
wine_data['volatile acidity'] = wine_data['volatile acidity'].fillna(mean_va)
```
